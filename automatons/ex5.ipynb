# 3 Polynomial regression

```python
def ploy_features(X, p):
    x_1 = X[1]
    for i in xrange(2, p+1):
        X = np.vstack((X, x_1**i))
    return X
```

## 3.1 Learning polynomial regression

```python
def feature_normalize(X):  
    X_poly = np.zeros_like(X)
    X_poly[0] = X[0]
    mu = np.mean(X[1:], 1)
    sigma = np.std(X[1:], 1)
    for i, feature in enumerate(X[1:]):
        feature = (feature - mu[i]) / sigma[i]
        X_poly[i+1] = feature
    return X_poly, mu, sigma
```

```python
def val_or_test_feature_normalize(X, mu, sigma):  
    X_poly = np.zeros_like(X)
    X_poly[0] = X[0]
    for i, feature in enumerate(X[1:]):
        feature = (feature - mu[i]) / sigma[i]
        X_poly[i+1] = feature
    return X_poly
```

## 3.5 Plotting learning curves with randomly selected examples

The exercise asks use to calculate cross validation error using i validation examples.  Here, instead, cross validation error is calculated using all validation examples.

```python
def learning_curve_rand_samples(lin_theta_0, X, y, X_val, y_val, lam, num_rand_samples):
    errors_train, errors_val = np.zeros((M-1, num_rand_samples)), np.zeros((M-1, num_rand_samples))
    for i in xrange(2, M+1):
        for j in xrange(num_rand_samples):
            rand_train_examples = np.random.randint(0, M, i)
            rand_val_examples = np.random.randint(0, M_val, i)
            res_train = train_linear_reg(lin_theta_0, X.T[rand_train_examples].T, y[rand_train_examples], lam)
            theta = res_train['x']
            error_train = lin_reg_cost_function(theta, X.T[rand_train_examples].T, y[rand_train_examples], 0)
            error_val = lin_reg_cost_function(theta, X_val, y_val, 0)
            errors_train[i-2, j], errors_val[i-2, j] = error_train, error_val
    return errors_train, errors_val
```

```python
lam = 0.01
num_rand_samples = 100
lin_theta_0 = np.ones(poly_deg+1)
errors_train, errors_val = \
    learning_curve_rand_samples(lin_theta_0, X_poly, y, X_poly_val, y_val, lam, num_rand_samples)
```

```python
plt.figure()
plt.errorbar(np.arange(2, M+1), np.mean(errors_train, axis=1), 
             np.std(errors_train, axis=1) / np.sqrt(num_rand_samples), # s.e.m. error bars
             c='b', ls='-', marker='o', mfc='w', mec='b', ms=6, label='Train')
plt.errorbar(np.arange(2, M+1), np.mean(errors_val, axis=1), 
             np.std(errors_val, axis=1) / np.sqrt(num_rand_samples), # s.e.m. error bars
             c='g', ls='-', marker='o', mfc='w', mec='g', ms=6, label='Cross validation')
plt.xlabel('Number of training examples')
plt.ylabel('Error')
plt.title('Polynomial regression learning curve ' r'($\lambda=%.2f$)' % lam)
plt.xlim(0, 13)
plt.ylim(-10, 200)
plt.legend(numpoints=1, loc=9)
plt.savefig('ex5_plots/fig_10.png', dpi=300)
```
